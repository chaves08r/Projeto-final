```python
# Criando tabela cruzada para gênero e faixa salarial
abs_genero = pd.crosstab(df_novo_2023["Faixa_salarial"], df_novo_2023["Genero"])

# Preparando dados para o gráfico
women_pop = list(abs_genero["Feminino"])
men_pop = [-1 * element for element in abs_genero["Masculino"]]  # Multiplicar por -1 para criar a pirâmide

faixa = list(abs_genero.index.values)

# Criando a figura do gráfico
fig = go.Figure()

# Adicionando barras para a população masculina
fig.add_trace(go.Bar(
    y=faixa,
    x=men_pop,
    name='Masculino',
    marker_color='lightgray',
    orientation='h'
))

# Adicionando barras para a população feminina
fig.add_trace(go.Bar(
    y=faixa,
    x=women_pop,
    name='Feminino',
    marker_color='#710c04',
    orientation='h'
))

# Atualizando o layout do gráfico
fig.update_layout(
    title="Pirâmide salarial segundo o gênero",
    yaxis_title='Faixa salarial',
    xaxis_title='Número de pessoas',
    barmode='relative',
    xaxis_tickangle=-90,
    plot_bgcolor='rgba(0,0,0,0)',
    width=800,
    height=500,
    legend=dict(title='Gênero')
)

# Exibindo o gráfico
fig.show()
```
